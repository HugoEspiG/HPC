## ¿Cuál es la exactitud del modelo?
___

¿Qué tantos errores comete nuestro clasificador?

La exactitud de un modelo mediante la matriz de confusion esta dada por la siguiente funcion:


##$Accuracy=\frac{TN + TP}{TN+FP+FN+TP}$


Remplazando los valores obtenidos obtenemos la formula:

##$\frac{4 + 3}{3 + 2 + 1 + 4}=0.7$



La exactitud del modelo es del $70\%$



##¿Cuál es la sensibilidad del modelo?
___

¿Qué proporción de los valores positivos reales fue clasificada correctamente? 

La sensibilidad de un modelo mediante la matriz de confusion esta dada por la siguiente funcion:


##$Recall=\frac{TP}{TP+FN}$


Remplazando los valores obtenidos obtenemos la formula:

##$\frac{3}{3+2}=0.6$



La sensibilidad del modelo es del $60\%$


##¿Cuál es la precisión del modelo?
___

¿Qué proporción de las clasificaciones positivas fueron correctas?

La precision de un modelo mediante la matriz de confusion esta dada por la siguiente funcion:


##$Precision=\frac{TP}{TP+FP}$


Remplazando los valores obtenidos obtenemos la formula:

##$\frac{3}{3+1}=0.75$



La precision del modelo es del $75\%$

##¿Cuál es el valor F1?
___

El valor $F_1$ de la matriz de confusion esta dado por alguna de las siguientes funciones:


##$F_1=\frac{2 \cdot precision \cdot recall}{precision + recall}$

##$F_1=\frac{TP}{TP+\frac{1}{2}(FP+FN)}$


Remplazando los valores obtenidos obtenemos la formula:

##$\frac{2 \cdot 0.75 \cdot 0.6}{0.75 + 0.6}=0.666$



El valor $F_1$ es de $60,66\%$


##¿Podríamos decir que este modelo es un buen sistema de recomendación?
___

Teniendo en cuenta las metricas de precision, el modelo presentado por el sitio web no tiene un buen sistema de recomendacion, ya que cuenta con un valor de exactitud del $70\%$ el cual determina la cantidad de errores que comete el clasificador, siendo este un porcentaje muy alto para la recomendacion del modelo. Adicionalmente presenta una sensibilidad del $60\%$ la cual denota los valores positivos que fueron clasificados correctamente, este valor sensibilidad es bajo, lo que proporciona una poca efectividad del modelo.
